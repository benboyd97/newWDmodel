```python
from astropy.io import ascii
from matplotlib import pyplot as plt
new = ascii.read("new_cal.dat")
old = ascii.read("WDphot_c25_ILAPHv5_abmag.dat")
```

```python
bands = ['F275W','F336W','F475W','F625W','F775W','F160W']
```

```python
for b in bands:

    fig, (ax_main, ax_sub) = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), gridspec_kw={'height_ratios': [3, 1]})
    ax_main.set_title('WD Magnitude Calibration '+b )
    ax_main.scatter(old[b][3:], new[b][3:], label='Main Plot',color='blue')
    #ax_main.errorbar(old[b][3:], new[b][3:],xerr=old['d'+b][3:], yerr=new['d'+b][3:],fmt='none',color='blue')
    ax_main.set_ylabel('Numpyro (mag)')

    ax_sub.scatter(old[b][3:], new[b][3:]-old[b][3:], label='Smaller Subplot', color='blue')
    #ax_sub.errorbar(old[b][3:], new[b][3:]-old[b][3:],xerr=old['d'+b][3:], yerr=new['d'+b][3:],fmt='none',color='blue')
    ax_sub.set_xlabel('Pymc (mag)')
    ax_sub.set_ylabel('Numpyro - Pymc')

    plt.tight_layout()


    plt.show()
```
